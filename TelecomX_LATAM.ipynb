```python
# Crear la columna Cuentas_Diarias
df.loc[:,'account.Charges.Daily'] = (df['account.Charges.Monthly'] / 30).round(2)
```

```python
df[['account.Charges.Daily', 'account.Charges.Monthly']].head(5)
```

Se cre贸 la columna Cuentas_Diarias dividiendo la facturaci贸n mensual entre 30 d铆as, proporcionando una m茅trica de gasto diario por cliente. Esta nueva variable permite analizar de manera m谩s granular el comportamiento de los clientes y detectar patrones de evasi贸n en funci贸n del gasto diario.

```python
df.head(5)
```

```python
for col in df.select_dtypes(include='object').columns:
    print(col, "", df[col].unique(), "\n")
```

```python
# Convertir valores textuales en binarios
binarias = ['customer.Partner', 'customer.Dependents', 'phone.PhoneService',
            'phone.MultipleLines', 'internet.OnlineSecurity', 'internet.OnlineBackup',
            'internet.DeviceProtection', 'internet.TechSupport', 'internet.StreamingTV',
            'internet.StreamingMovies', 'account.PaperlessBilling']
```

```python
# se transforma
for col in binarias:
    df.loc[:,col] = df[col].map({'Yes': 1, 'No': 0,'No internet service': 0, 'No phone service': 0 })
```

```python
df.sample(5)
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
import numpy as np

# Suponiendo que tu columna de gasto diario se llama 'Cuentas_Diarias'
# Verificamos m铆nimos y m谩ximos
min_val = df['account.Charges.Daily'].min()
max_val = df['account.Charges.Daily'].max()
```

```python
# Creamos 5 bins equidistantes entre min y max
bins = np.linspace(min_val, max_val, 6)  # 6 l铆mites = 5 segmentos
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
```

```python
# Crear columna segmentada
df['Daily_Segment'] = pd.cut(
    df['account.Charges.Daily'],
    bins=bins,
    labels=labels,
    include_lowest=True  # Incluye el valor m铆nimo en el primer bin
)
```

```python
# Verificamos el resultado
print(df[['account.Charges.Daily', 'Daily_Segment']].head(10))
print("\nDistribuci贸n de segmentos:")
print(df['Daily_Segment'].value_counts())
```
