A partir de esta revisi贸n, se identific贸 la variable Churn como objetivo, y se seleccionaron como variables relevantes aquellas relacionadas con el perfil del cliente, los servicios contratados y las condiciones del contrato.

```python
df = datos_TelecomX_normalizados.copy()
```

```python
for col in df.select_dtypes(include="object").columns:
  df[col] = df[col].str.strip()
```

```python
df['Churn'].unique()
```

```python
churn_vacios = df[df['Churn'] == '']
```

```python
churn_vacios
```

```python
churn_vacios[['customer.tenure', 'account.Charges.Total', 'account.Charges.Monthly']].head()
```

```python
# eliminar los registros con df['Churn'] vacios
df = df[df['Churn'] != '']
```

```python
# df['Churn'] = df['Churn'].str.capitalize()
```

```python
df['Churn'].unique()
```

Se identificaron 224 registros con la variable objetivo Churn vac铆a. Al explorar estos clientes se observ贸 que poseen historial de facturaci贸n y antig眉edad variable (customer.tenure), pero no es posible inferir su estado de evasi贸n. Por este motivo, estos registros fueron eliminados del conjunto de datos, asegurando la integridad de la variable objetivo.

```python
df['account.Charges.Total'].dtype
```

```python
df['account.Charges.Total']= pd.to_numeric(df['account.Charges.Total'], errors='coerce')
```

```python
df['account.Charges.Total'].dtype
```

```python
df.isnull().sum()
```

```python
df.loc[df['account.Charges.Total'].isnull(), ['account.Charges.Total', 'customer.tenure', 'account.Charges.Monthly']]
```

```python
df.loc[df['customer.tenure'] == 0, 'account.Charges.Total'] = 0
```

```python
df['account.Charges.Total'].isnull().sum()
```

Se identificaron 11 valores nulos en la variable account.Charges.Total. Al analizar estos registros se comprob贸 que corresponden a clientes con antig眉edad (customer.tenure) igual a cero, quienes a煤n no han generado cargos acumulados. Por este motivo, los valores nulos fueron actualizados con el valor 0, garantizando la coherencia de la informaci贸n financiera.
